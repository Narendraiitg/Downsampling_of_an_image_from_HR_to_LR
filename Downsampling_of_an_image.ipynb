```python
# Imporing the various libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/HR_image.jpg") # Reading the HR input image

print("The dimension of original img",img.shape) # Printing the dimention of HR input image
cv2_imshow(img) # Dispalying the HR input image

# Resizing the HR input image from 500 X 1200 to 400 X 600
img_re= cv2.resize(img, (600,400))

print("Dimensions after reshaping",img_re.shape)# Printing the dimention of resized image
cv2_imshow(img_re)# Dispalying the resized image
```

```python
layer= img_re.copy() # Creating the copy of resized image
img_1=cv2.pyrDown(layer) # Performing the downsampling by 2 on resized image
img_2=cv2.pyrDown(img_1) # Performing the downsampling by 2 on downsampled image

cv2_imshow(layer)#This function displays the copy of resized image.
print("dimension of original resized image",layer.shape) # Dimensions of copy of resized image

cv2_imshow(img_1) #This will display downsampled version image of resized image by factor of 2.
print("dimension of LR image downsampled x2",img_1.shape) # Printing the dimension of downsampled image by factor of 2

cv2_imshow(img_2) #This will display downsampled version image of resized image by factor of 4.
print("dimension of LR downsampled image x4",img_2.shape) # Printing the dimension of downsampled image by factor of 4
cv2.waitKey(0)
cv2.destroyAllWindows()
```
